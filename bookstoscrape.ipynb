```python
# Importing the Package
import requests
from bs4 import BeautifulSoup
import pandas as pd

book_list = []

for page_number in range(51):
    url = f'http://books.toscrape.com/catalogue/page-{page_number}.html'

    r = requests.get(url)

    soup = BeautifulSoup(r.text, 'lxml')


    article = soup.find_all('article', class_='product_pod')

    for book in article:
        title = book.find_all('a')[1]['title']
        price = book.find('p', class_='price_color').text[2:]
        instock = book.find('p', class_='instock availability').text.strip()
        books = {
            'Title': title,
            'Price': price,
            'In-Stock':instock
        }

        book_list.append(books)
```
