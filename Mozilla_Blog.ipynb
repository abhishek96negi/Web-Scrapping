```python
#Scrape blog posts with Python

import requests
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd
article_list = []
x = 1
while True:
  #  try:
        url = f'https://blog.mozilla.org/en/category/mozilla/news/page/{x}/'

        headers = {
            'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'
        }

        r = requests.get(url, headers=headers)
        r
        soup = BeautifulSoup(r.content, features='lxml')
        soup
        articles = soup.find_all('section', class_='mzp-c-card mzp-has-aspect-1-1')
        if len(articles) == 0:
            break

        for article in articles:
            image = article.img['src'] if article.img is not None else article.find('div', class_='ft-c-fallback')['style'].split('(')[1].split(')')[0]
            tag = article.find('div', class_='ft-c-pill').text
            title = article.find('h2', class_='mzp-c-card-title').text
            link = article.a['href']
            article_dict = {
                'title': title,
                'tag': tag,
                'image': image,
                'link': link
            }
            article_list.append(article_dict)
        sleep(3)
        x+= 1
   # except Exception as e:
    #    print(e)
   #     break
```

```python
article_df = pd.DataFrame(article_list)
article_df
```
