```python
import requests
from bs4 import BeautifulSoup

url = 'https://store.steampowered.com/search/results/?query&start=0&count=50&dynamic_data=&sort_by=_ASC&snr=1_7_7_7000_7&filter=topsellers&tags=19&infinite=1'
r = requests.get(url)
r
```

```python
game_list = []
for x in range(0, r.json()['total_count'], 50):
    url = f'https://store.steampowered.com/search/results/?query&start=0&count={x}&dynamic_data=&sort_by=_ASC&snr=1_7_7_7000_7&filter=topsellers&tags=19&infinite=1'
    r = requests.get(url)
    soup = BeautifulSoup(r.json()['results_html'], 'lxml')
    games = soup.find_all('a', {'class': 'search_result_row'})
    for game in games:
        game_url = game['href']
        title = game.find('span', {'class': 'title'}).text
        date = game.find('div', {'class': 'search_released'}).text
        price = game.find('div', {'class': 'search_price'}).text.strip()
        game_dict = {
            'URL': game_url,
            'Name': title,
            'Date': date,
            'Price': price
        }
        game_list.append(game_dict)
```

```python
import pandas as pd
df = pd.DataFrame(game_list)
df
```
